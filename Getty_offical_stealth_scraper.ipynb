```python
from curl_cffi import requests
from bs4 import BeautifulSoup

player_images = []

def stealth_scrape(entity):
    url = f"https://www.gettyimages.com/photos/{entity.replace(' ', '-')}"

    # curl_cffi mimics the TLS fingerprint of a real browser
    response = requests.get(url, impersonate="chrome120")

    if response.status_code == 200:
        soup = BeautifulSoup(response.content, "html.parser")
        # Getty images are often in <img> tags with 'MosaicAsset' in the class
        images = soup.find_all('img')
        for img in images[:20]:
            player_images.append(img.get('src'))
            # print(img.get('src'))
    else:
        print(f"Blocked! Status Code: {response.status_code}")
#_______________________________________
entity = "Alejandro Rego"
query = "Alejandro Rego Portrait"
#_______________________________________

stealth_scrape(query)

# Splitting entity to check for individual keywords (case-insensitive)
keywords = [k.lower() for k in entity.split()]

filtered_images = [
    link for link in player_images
    if all(word in link.lower() for word in keywords)
]

print(f"Total images found: {len(filtered_images)}")
for link in filtered_images:
    print(link)
```
